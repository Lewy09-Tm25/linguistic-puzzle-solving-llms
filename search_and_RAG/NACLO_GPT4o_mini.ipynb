```python
def gpt_run(bases):
  img_urls = [{"type": "image_url", "image_url": {"url": f"data:image/png;base64,{base}"}} for base in bases]
  response = client.chat.completions.create(
      model=MODEL,
      messages=[
          {"role": "system", "content": "You are a helpful assistant that responds in Markdown. Help me with the questions!"},
          {"role": "user", "content": [
              {"type": "text", "text": "Solve all the questions in the file."}
          ]+img_urls}
      ],
      temperature=0.0,
  )
  return response.choices[0].message.content
```

```python
# Open the image file and encode it as a base64 string
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
```

```python
from IPython.display import Image, display, Audio, Markdown
import base64, os

# Path to the folder containing all images
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2023_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2022_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2021_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2020_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2019_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2018_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2017_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2016_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2015_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2014_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2013_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2012_pdf_ques_imgs/"
# IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2011_pdf_ques_imgs/"
IMAGE_FOLDER = "/content/gdrive/My Drive/Sypdfs/img_questions_pdfs/2010_pdf_ques_imgs/"

# List to store messages for each PDF
messages = []

# Iterate over each image file associated with the PDF
page_number = 1
prev_lang = None
bases = []
# Iterate over each PDF file in the folder
for i,pdf_file in enumerate(os.listdir(IMAGE_FOLDER)):

    # Get the name of the PDF file without extension
    pdf_name = os.path.splitext(pdf_file)[0]
    # Construct the image path correctly
    image_path = os.path.join(IMAGE_FOLDER, f"{pdf_name}.png")
    # Get the name
    lang = (' '.join(pdf_file.split(' ')[1:])).split('.')[0]
    if prev_lang != lang:
      if prev_lang is not None:
        print(f"LANGUAGE:- {prev_lang}->")
        res = gpt_run(bases)
        print(res)
        print("----------------")
        bases = []
      prev_lang = lang
      if i == len(os.listdir(IMAGE_FOLDER)) - 1:
        base64_image = encode_image(image_path)
        bases.append(base64_image)
        print(f"LANGUAGE:- {prev_lang} ->")
        res = gpt_run(bases)
        print(res)
        print("----------------")
    elif prev_lang == lang and i == len(os.listdir(IMAGE_FOLDER)) - 1:
      base64_image = encode_image(image_path)
      bases.append(base64_image)
      print(f"LANGUAGE:- {prev_lang} ->")
      res = gpt_run(bases)
      print(res)
      print("----------------")
    base64_image = encode_image(image_path)
    bases.append(base64_image)
```
