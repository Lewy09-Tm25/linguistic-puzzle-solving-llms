```python
from PyPDF2 import PdfReader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI

# Change the file path accordingly
pdf_reader = PdfReader("/content/gdrive/MyDrive/Sypdfs/pdfs/2014_pdfs/2014_questions/2014_R2_3_Yidiny_1_3.pdf")
text = ""
for page in pdf_reader.pages:
    text += page.extract_text()

# Split into chunks
text_splitter = CharacterTextSplitter(
    separator="\n", # Defines a new line
    chunk_size = 1000,
    chunk_overlap = 200,
    length_function = len
)
chunks = text_splitter.split_text(text)

# Create embeddings
embeddings = OpenAIEmbeddings()
# Creating an object on which we will be able to search FAISS
knowledge_base = FAISS.from_texts(chunks, embeddings)
```

```python
# Prompt
user_question = "Understand the whole text, and solve the questions."
docs = knowledge_base.similarity_search(user_question)

llm = OpenAI()
chain = load_qa_chain(llm, chain_type="stuff")
response = chain.run(input_documents=docs, question = user_question)
```
