```python
import os
import fitz

path = "pdfs/2024_pdfs/2024_questions/"
for filename in os.listdir(path):
    if filename.endswith(".pdf"):
        input_path = os.path.join(path, filename)
        doc = fitz.open(input_path)
        zoom = 4
        mat = fitz.Matrix(zoom, zoom)
        count = 0
        # Count variable is to get the number of pages in the pdf
        for p in doc:
            count += 1
        for i in range(count):
            val = f"2024 pdf ques to imgs/{filename}_image_{i+1}.png"
            page = doc.load_page(i)
            pix = page.get_pixmap(matrix=mat)
            pix.save(val)
        doc.close()
```

```python
import os
import fitz

for year in range(2010, 2024):
    path = f"pdfs/{year}_pdfs/{year}_questions/"
    path_to_save = f"{year}_pdf_ques_imgs/"
    os.mkdir(path_to_save)
    for filename in os.listdir(path):
        if filename.endswith(".pdf"):
            input_path = os.path.join(path, filename)
            doc = fitz.open(input_path)
            zoom = 4
            mat = fitz.Matrix(zoom, zoom)
            count = 0
            # Count variable is to get the number of pages in the pdf
            for p in doc:
                count += 1
            for i in range(count):
                fname = f"{filename}_image_{i+1}.png"
                output_path = os.path.join(path_to_save, fname)
                page = doc.load_page(i)
                pix = page.get_pixmap(matrix=mat)
                pix.save(output_path)
            doc.close()
        print(f"{filename} DONE")
    print(f"-----------{year} DONE")
```
